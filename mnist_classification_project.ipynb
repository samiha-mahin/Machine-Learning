```python
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml("mnist_784", version=1, as_frame=False)
print(mnist)
```

```python
import pandas as pd
mnist_show = pd.DataFrame(mnist.data)
mnist_show.head(5)
```

```python
x = mnist['data']
y = mnist['target']
x_train, x_test, y_train, y_test = x[:60000], x[60000:], y[:60000], y[60000:]
```

```python
x_test
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
display = x[2]   # Selecting the n'th image.like: 5000th image, or here 0 th image
display_image = display.reshape(28, 28) # Reshaping it to get the 28x28 pixels
plt.imshow(display_image, cmap=matplotlib.cm.binary, interpolation="nearest")
#plt.axis("off")
plt.show()
```

```python
y[2]
```

```python
from sklearn import tree
dtc = tree.DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

```python
dtc.predict(x_test)
```

```python
dtc.score(x_test, y_test)
```

```python
from sklearn import metrics

# Predict the values for the test set and store them in a variable called 'predict'
predict = dtc.predict(x_test)

# Now use the 'predict' variable in the classification_report function
print(metrics.classification_report(y_test, predict))
```

```python
print(metrics.confusion_matrix(y_test, predict))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
#prediction
predict = lr.predict(x_test)
#print(predict)
lr.score(x_test, predict)
```

```python
#prediction
predict = lr.predict(x_test)
#print(predict)
lr.score(x_test, predict)

print(metrics.classification_report(y_test, predict)) # Changed 'paredict' to 'predict'
```
