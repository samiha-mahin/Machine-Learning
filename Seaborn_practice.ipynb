```python
import seaborn as sns
import pandas as pd
from IPython.display import display
```

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
#distribution plot
sns.histplot(data=df, x='bill_length_mm',kde=True, hue='island')
```

```python
#joint plot
sns.jointplot(data=df, x='bill_length_mm', y='flipper_length_mm', hue='island')
```

```python
sns.jointplot(data=df, x='bill_length_mm', y='flipper_length_mm', kind='reg')
```

```python
#pairplot
sns.pairplot(data=df)
```

```python
sns.countplot(data=df, x='island',palette='PuBu')
```

```python
sns.barplot (data=df, y='flipper_length_mm', x='sex',hue='island',palette='PuBu')
```

```python
sns.boxplot(data=df, y='flipper_length_mm', x='island',hue='species',palette='PuBu')
```

```python
sns.violinplot(data=df, y='flipper_length_mm', x='island',hue='species',palette='PuBu')
```

```python
#pairgrid
g = sns.PairGrid(df)
g.map(sns.scatterplot)
g.map_diag(sns.histplot)
```
